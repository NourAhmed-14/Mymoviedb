```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
## remove warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv('/content/mymoviedb.csv', lineterminator='\n')
movies.head()
```

**Dropping the poster url column, i won't be using it**

```python
movies.drop(['Poster_Url'],axis=1,inplace=True)
movies.head(1)
```

**Converting the type of realese date fro object to datetime**

```python
movies['Release_Date']=pd.to_datetime(movies['Release_Date'])
movies['Release_Date'].info()
```

```python
movies['Release_year']=movies['Release_Date'].dt.year
```

```python
movies.drop(['Release_Date'],axis=1,inplace=True)
```

```python
movies['Release_year']=movies['Release_year'].astype(int)
#movies.drop(['Release_yeat'],axis=1,inplace=True)
movies.head(1)
```

# What is the movie with the highest Popularity rate? And what kind is it?

```python
top_popular= movies.sort_values(by='Popularity',ascending=False).head(10)

plt.figure(figsize=(15,8))
sns.set(style='darkgrid')
sns.barplot(y='Popularity',x='Title',data=top_popular,palette='rocket')
plt.xticks(rotation=90, fontsize=10)
plt.title('Top 10 popular movies',fontweight='bold',backgroundcolor='#b9778b',color='white')
plt.xlabel('Movie Title')
plt.ylabel('Popularity')
plt.show()
```

**The genre of top 10 popular movies**

```python
top_popular[['Title','Genre']]
```

# What year was the largest number of films produced?

```python
movies_per_year=movies['Release_year'].value_counts().sort_index().reset_index()
movies_per_year.columns=['year','movies_per_year']

plt.figure(figsize=(10,6))
sns.set(style='darkgrid')
sns.lineplot(x='year',y='movies_per_year',data=movies_per_year,color='#b74944')
plt.xticks(rotation=90, fontsize=10)
plt.title('Number of movies per year',fontweight='bold',backgroundcolor='#c87a76',color='white')
```

```python
movies_per_year=movies['Release_year'].value_counts().sort_index().reset_index()
movies_per_year.columns=['year','movies_per_year']

plt.figure(figsize=(10,6))
sns.set(style='darkgrid')
sns.scatterplot(x='year',y='movies_per_year',data=movies_per_year,color='#a14a45')
```

# What is most of the film genre?

```python
most_genre=movies['Genre'].value_counts().reset_index().head(30)
most_genre.columns = ['Genre', 'Count']

plt.figure(figsize=(15,8))
sns.set(style='darkgrid')
sns.barplot(x='Genre',y='Count',data=most_genre,palette="ch:s=-.2,r=.6")
plt.xticks(rotation=90, fontsize=10)
plt.title('Most genre',fontweight='bold',backgroundcolor='#b9778b',color='white')
plt.xlabel('Number of movies')
plt.ylabel('Genre')
plt.show()
```

# What kind of movies take the highest rating?

```python
most_rating=movies.sort_values(by='Vote_Average',ascending=False).head(20)

plt.figure(figsize=(10,6))
sns.set(style='darkgrid')
sns.barplot(y='Vote_Average',x='Title',data=most_rating,palette='crest')
plt.xticks(rotation=90, fontsize=10)
plt.title('Top 20 rating movies',fontweight='bold',backgroundcolor='#60a69a',color='white')
plt.xlabel('Movie Title')
plt.ylabel('Vote_Average')
```

```python
most_rating[['Title','Genre']]
```

# Which movie is more popular? Which movie is less popular?

```python
most_popular_movie=movies.loc[movies['Popularity'].idxmax()]
least_popular_movie=movies.loc[movies['Popularity'].idxmin()]

popularity_df=pd.DataFrame([most_popular_movie,least_popular_movie])

plt.figure(figsize=(10,6))
sns.set(style='darkgrid')
sns.lineplot(x='Popularity',y='Title',data=popularity_df,color='#712929')
plt.title('Most and least popular movies',fontweight='bold',backgroundcolor='#653d3d',color='white')
```

```python
movies_sort=movies.sort_values(by='Popularity',ascending=False)
head_tail_df=pd.concat([movies_sort.head(10), movies_sort.tail(10)])

plt.figure(figsize=(10,6))
sns.set(style='darkgrid')
sns.scatterplot(x='Popularity',y='Title',data=head_tail_df,color='#189390')
plt.title('Top 10 popular movies & Least 10 popular',fontweight='bold',backgroundcolor='#0c514f',color='white')
```
